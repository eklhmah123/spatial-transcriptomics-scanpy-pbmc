```python
from spatialdata_io import xenium
import spatialdata as sd

# Define sdata from your Xenium_Lung_Cancer directory
xenium_path = "./Xenium_Lung_Cancer"
sdata = xenium(xenium_path, cells_as_circles=True)

print("sdata is now defined!")
print(sdata)
```

```python
adata = sdata.tables["table"]
adata
```

```python
adata.obs
```

```python
adata.obsm["spatial"]
```

```python
sc.pp.calculate_qc_metrics(adata, percent_top=(10, 20, 50, 150), inplace=True)
```

```python
cprobes = (
    adata.obs["control_probe_counts"].sum() / adata.obs["total_counts"].sum() * 100
)
cwords = (
    adata.obs["control_codeword_counts"].sum() / adata.obs["total_counts"].sum() * 100
)
print(f"Negative DNA probe count % : {cprobes}")
print(f"Negative decoding count % : {cwords}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(1, 4, figsize=(15, 4))

axs[0].set_title("Total transcripts per cell")
sns.histplot(adata.obs["total_counts"], kde=False, ax=axs[0])

axs[1].set_title("Unique transcripts per cell")
sns.histplot(adata.obs["n_genes_by_counts"], kde=False, ax=axs[1])

axs[2].set_title("Area of segmented cells")
sns.histplot(adata.obs["cell_area"], kde=False, ax=axs[2])

axs[3].set_title("Nucleus ratio")
sns.histplot(adata.obs["nucleus_area"] / adata.obs["cell_area"], kde=False, ax=axs[3])

plt.tight_layout()
plt.show()
```

```python
sc.pp.filter_cells(adata, min_counts=10)
sc.pp.filter_genes(adata, min_cells=5)
```

```python
from spatialdata_io import xenium

# Use the path where experiment.xenium was found
sdata = xenium("./Xenium_Lung_Cancer", cells_as_circles=True)
print("Success!")
```

```python
# Get adata from sdata
adata = sdata.tables["table"]

# Now create subsample
sdata.tables["subsample"] = sc.pp.subsample(adata, fraction=0.5, copy=True)
```

```python
adata_subsample = sdata.tables["subsample"]
```

```python
import scanpy as sc

# First compute neighbors
sc.pp.neighbors(adata_subsample)

# Then run leiden
sc.tl.leiden(adata_subsample)

# Now run co-occurrence
import squidpy as sq

sq.gr.co_occurrence(adata_subsample, cluster_key="leiden")
sq.pl.co_occurrence(adata_subsample, cluster_key="leiden", clusters="0", figsize=(10, 10))
sq.pl.spatial_scatter(adata_subsample, color="leiden", shape=None, size=2)
```
